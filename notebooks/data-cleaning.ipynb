```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Data
df = pd.read_csv("../data/raw-data.csv", sep=';')
print(df.head())
```

```python
# Initial Data Exploration
print(df.shape)
print(df.info())
```

```python
# Missing Values
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)

# No missing values found
```

```python
# Duplicate Rows
duplicates = df.duplicated().sum()
print("Duplicate rows:", duplicates)

# No duplicate rows found
```

```python
# We are predicting whether a student will dropout or not 
# so the number of "Enrolled" student is irrelevant
# We only need to know whether a student graduated or dropedout. 
# So we are dropping the "Enrolled" values and going forward with "Graduate" & "Dropout" values.

df.drop(df[df['Target'] == 'Enrolled'].index, inplace = True)
```

```python
# Converting Target Row to Numeric (for Heatmap)
df['Target'] = df['Target'].str.strip().map({'Dropout': 0, 'Graduate': 1})
```

```python
# Updated Data Exploration
print(df.shape)
print(df.info())
```

```python
# cleaned data exploration
df.head()
```

```python
# save to data folder
df.to_csv("../data/cleaned-data.csv", index=False)
```
