```python
# USING PANDAS-PROFILING TO GENERATE REPORT

from ydata_profiling import ProfileReport
import pandas as pd

DATA_PATH = "../../data/cleaned-data.csv"
df = pd.read_csv(DATA_PATH)
profile = ProfileReport(
    df,
    title="Student Dropout Risk - Data Profiling Report",
    explorative=True
)

profile.to_file("../EDA Results/eda_pandas_profiling.html")
```

```python
# USING AUTOVIZ TO GENERATE VISUALIZATIONS

from autoviz.AutoViz_Class import AutoViz_Class
import pandas as pd

DATA_PATH = "../../data/cleaned-data.csv"
df = pd.read_csv(DATA_PATH)
AV = AutoViz_Class()

target_col = "Target"

AV.AutoViz(
    filename="",
    dfte=df,
    depVar=target_col,
    verbose=2,
    lowess=False,
    chart_format="png",
    max_rows_analyzed=5000,
    max_cols_analyzed=50
)
```
